```python
import pandas as pd
import numpy as np



import seaborn as sns
```

```python
data = pd.read_csv("DailyClimateTrain.csv")
```

```python

forecast = data.rename(columns={"date":"ds","meantemp":"y"})
```

```python
forecast
```

```python

from prophet import Prophet
```

```python
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
model = Prophet()
```

```python
model.fit(forecast)
```

```python
forecast_pred = model.make_future_dataframe(periods=3650)
```

```python
pred=model.predict(forecast_pred)
```

```python
plot_plotly(model,pred)
```
